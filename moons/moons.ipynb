```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

[width, height] = plt.rcParams['figure.figsize']
plt.rcParams['figure.figsize'] = [width*3, height*2]
```

```python
data = np.loadtxt('moons.txt', dtype='int')
step = data[:,0]
p = []
v = []
for moon in range(4):
    idx = moon * 2
    p.append(data[:,idx+1])
    v.append(data[:,idx+2])
```

```python
for moon in range(4):
    plt.plot(step, p[moon])
    plt.plot(step, v[moon])
    plt.show()
```

```python
total_v = np.sum(v, axis=0)
abs_v = np.absolute(v)
total_abs_v = np.sum(abs_v, axis=0)
plt.plot(step, total_v)
plt.plot(step, total_abs_v)
plt.show()
```
